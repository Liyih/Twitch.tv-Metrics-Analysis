```python
#Testing scraper for Twitchtracker
import urllib.request as urllib
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

import pandas as pd

#bypass 403 forbidden against web crawlers
req = Request('https://twitchtracker.com/statistics', headers={'User-Agent': 'Mozilla/5.0'})
html = urlopen(req).read()

soup = BeautifulSoup(html)

table = soup.find(id="big-one");
rows = table.find_all("tr");

df = pd.DataFrame(columns = ["Month", "Avg. Concur. Viewers", "Avg. Concur. Channels", "Time Watched", "Active Streamers"])

for row in rows:
    data = row.find_all("td");
    
    #Knowing the structure of the table, process all valid rows in standard order
    #Append each processed data item to our dataframe
    if len(data) != 0:
        month = data[0].get_text().strip()
        viewers = data[1].get_text().strip()
        channels = data[2].get_text().strip()
        watched = data[3].get_text().strip().split(" ")[0]
        streamers = data[4].get_text().strip()
#       If you want to see individual outputs
#         print("Date: " + month)
#         print("Viewers: " + viewers)
#         print("Channels: " + channels)
#         print("Time Watched: " + watched)
#         print("Streamers: " + streamers)
        data_dict = {"Month" : month, "Avg. Concur. Viewers": viewers, 
                     "Avg. Concur. Channels": channels, "Time Watched": watched, "Active Streamers": streamers}
        df = df.append(data_dict, ignore_index=True)
        
#Write our data and display the data
df.to_csv("totaltwitchdata.csv")
df

```
