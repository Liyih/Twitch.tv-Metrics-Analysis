```python
#Aggregating Downloaded CSV Files for Twitch

import pandas as pd

#Masterdf holds all data
#Df is the dataframe to append
#Month is a string to track month
def dfprocess(masterdf, df, month):
    df = df.drop(df.columns[[0, 1, 13]], axis=1)
    df = df.set_index("Channel");
    df["Month"] = month;
    masterdf = masterdf.append(df);
    return masterdf;  

#Initialize our dataframe to the latest month
dfmaster = pd.read_csv("data/April2020.csv");
dfmaster = dfmaster.drop(dfmaster.columns[[0, 1, 13]], axis=1)
dfmaster = dfmaster.set_index("Channel");
dfmaster["Month"] = "4-2020";

csvarray = ["March2020.csv", "February2020.csv", "January2020.csv", "December2019.csv", 
            "November2019.csv", "October2019.csv", "September2019.csv", "August2019.csv", 
            "July2019.csv", "June2019.csv", "May2019.csv", "April2019.csv"];
montharray = ["2020-03-01", "2020-02-01", "2020-01-01", "2019-12-01", "2019-11-01", "2019-10-01", "2019-09-01", 
             "2019-08-01", "2019-07-01", "2019-06-01", "2019-05-01", "2019-04-01"]

for i in range(len(csvarray)):
    dftemp = pd.read_csv("data/" + csvarray[i])
    dfmaster = dfprocess(dfmaster, dftemp, montharray[i])
    
dfmaster = dfmaster.sort_values(by=["Channel", "Watch time"], ascending=[True, False]);

# Counts the occurances of a certain channel
channelcount = dfmaster.index.value_counts()

# Adds all channel names that appear less than 10 times in the dataset to the list "invalidchannels"
invalidchannels = []
mintimes = 10
for i in range(len(channelcount)):
    if channelcount[i] < mintimes:
        invalidchannels.append(channelcount.index[i])

#Drops all rows with the channel names in the list "invalidchannels" from the Dataframe 
for value in invalidchannels:
    dfmaster.drop(value, inplace=True)

indexes = dfmaster.index 

for value in dfmaster.index:
    first_val = value.find("(") + 1
    foreign_name = value
    if first_val != 0:
        second_val = value.find(")")
        foreign_name = value[first_val:second_val]
        x = value
    dfmaster = dfmaster.rename(index = {value : foreign_name})
    
    
counter = dfmaster.index.value_counts()
print("There are",counter.count(),"channels that appear at least",mintimes,"times in the dataset.")

    
dfmaster.to_csv("TotalChannelData.csv")                       
dfmaster             
```
